```python
import re
from collections import deque
```

```python
with open("zimand_new.tex", encoding="utf8") as file:
    data = file.read()
```

```python
variables={}
```

```python
start_queue=[m.start() for m in re.finditer('<<symbol>>', data)]
end_queue=deque([m.start() for m in re.finditer('<</symbol>>', data)])
variables={}
while(len(end_queue)!=0):
    i=0
    
    for item in start_queue:
        if(not(item<end_queue[0])):
            #print(start_queue.index(item)-1)
            start_index=start_queue.pop(i-1)
            
            end_index=end_queue.popleft()
            print(data[start_index:end_index].replace("<<symbol>>",'').replace("<</symbol>>",''))
            print("StartIndex:{0}, end_index:{1}".format(start_index, end_index))
            variable=data[start_index:end_index].replace("<<symbol>>",'').replace("<</symbol>>",'').strip()
            
            if(not(variable in variables) and len(variable)>0):
                variables[variable]=[]
                variables[variable].append([start_index, end_index])
            elif(len(variable)>0):
                variables[variable].append([start_index, end_index])
            break
        i+=1
    
    if(i==len(start_queue)):
        break

while(len(start_queue)!=0):
    start_index=start_queue.pop()
    end_index=end_queue.popleft()
    print(data[start_index+len("<<symbol>>"):end_index].replace("<<symbol>>",'').replace("<</symbol>>",''))
    print("StartIndex:{0}, end_index:{1}".format(start_index, end_index))
    variable=data[start_index:end_index].replace("<<symbol>>",'').replace("<</symbol>>",'').strip()
    if(not(variable in variables) and len(variable)>0):
        variables[variable]=[]
        variables[variable].append([start_index, end_index])
    elif(len(variable)>0):
        variables[variable].append([start_index, end_index])
```

```python
def replaceTags(text):
    from functools import reduce
    repls = ('<<symbol>>', ''), ('<</symbol>>', ''),('<<equation>>',''),('<</equation>>',''),('$','')
    text_update=reduce(lambda a, kv: a.replace(*kv), repls, text)
    return text_update
```

```python
#Sample Test
print(replaceTags("<<equation>>$<<symbol>>\sigma<</symbol>>$<</equation>>"))
```

```python
def findSentenceStart(data,index):
    given_index=index
    tags=['\end{abstract}', '\section{Introduction}', 'begin{abstract}']
    indexes=[]
    for item in tags:
        
        item_index=data.find(item)
        if(item_index>=0):
            indexes.append(item_index+len(item))
    print("All indexes:{}".format(indexes))
    #For finding the start of the sentences
    while(index>0):
        if(data[index]=='.'):
            break
        index-=1
        
    if(index>0):
        indexes.append(index+1)
    else:
        return indexes.append(0)
    
    #TO check out of tags and start of the sentences, what is closer to the given index
    min_=100000000
    req_index=0
    for sent_start_index in indexes:
        diff = given_index - sent_start_index
        print("Difference:{0}, given_index:{1}, sent_start_index:{2}".format(diff, given_index, sent_start_index))
        if(diff<min_ and diff>0):
            min_=diff
            req_index=sent_start_index
            #print("Yoyo"req_index)
    print("Req Index:{}".format(req_index))
    return req_index
```

```python
check_threshold=2 #This will pick the first two definitions
var_definitions={}
for key in variables: #Note that key is the variable name
    i=0
    for index in variables[key]: 
        start_index=findSentenceStart(data,index[0])
        #start_index=index[0]
        definition=data[start_index:].split('.')[0]
        print(definition)
        definition=replaceTags(definition)
        if(not(key in var_definitions)):
            var_definitions[key]=[]
        var_definitions[key].append(definition.strip())
        i+=1
        if(i>=check_threshold):
            break
        #print('key:{0}, definition{1}'.format(key, definition))
    
        
```
